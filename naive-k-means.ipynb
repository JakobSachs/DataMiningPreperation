```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_biclusters
```

```python
data = make_biclusters(shape=(200, 2), n_clusters=2, noise=5, random_state=0)[0]
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
def get_assignment(m_s,x):
  return np.argmin(np.linalg.norm(x-m_s,axis=1))

def get_assignments(m_s,x_s):
  return [get_assignment(m_s,x) for x in x_s]

def update_means(x_s,S_s):
  return [ np.mean([x for i,x in enumerate(x_s) if S_s[i] == k  ],axis=0) for k in range(K)]
```

```python
K = 2

# randomly pick means
rng = np.random.default_rng(2)
means = rng.choice(data,K)
orig_means = means


for _ in range(100):
  means = update_means(data,get_assignments(means,data))
```

```python
plt.scatter(data[:,0],data[:,1],c=get_assignments(orig_means,data))
plt.title("K-means with random centroids")
```

```python
plt.scatter(data[:,0],data[:,1],c=get_assignments(means,data))
plt.title("K-means with 100 iterations")
```
